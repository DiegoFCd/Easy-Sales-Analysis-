## **Modelo de regresion lineal (*statsmodel*)**<br>
statesmodel.png

```python
x = data['Vta A√±o anterior']
y = data['Venta REMA ($)']


x1 = sm.add_constant(x)
model = sm.OLS(y, x1).fit()
print(model.summary())

y_pred = model.predict(x1)
```

```python
# Hacer una copia expl√≠cita del DataFrame
data = data.copy()

# Agregar la nueva columna
data['PredictedVentaREMA ($)'] = y_pred

# Mostrar las primeras filas del DataFrame
data.head()
```

```python
def millones_formatter(x, pos):
    return f"{x:.1f}M"

def plot(x, y, y_pred):
  plt.scatter(x, y, color='red', label="Actual Venta REMA ($)")
  plt.scatter(x, y_pred, color='black',label="Predicted Venta REMA ($)")
  plt.plot(x, y_pred, color='green', label="L√≠nea de regresi√≥n", linewidth=2)


  # Formatear el eje Y en millones
  ax = plt.gca()
  ax.yaxis.set_major_formatter(ticker.FuncFormatter(millones_formatter))


  # A√±adir anotaciones para valores m√°ximos y m√≠nimos
  max_real_idx = y.idxmax()
  min_real_idx = y.idxmin()
  plt.annotate(f"M√°ximo: {y[max_real_idx]/1_000_000:.1f}M",
                 (x[max_real_idx], y[max_real_idx]),
                 textcoords="offset points", xytext=(0,10), ha='center', fontsize=9)
  plt.annotate(f"M√≠nimo: {y[min_real_idx]/1_000_000:.1f}M",
                 (x[min_real_idx], y[min_real_idx]),
                 textcoords="offset points", xytext=(0,-15), ha='center', fontsize=9)

  # Etiquetas y t√≠tulo
  plt.xlabel("Vta A√±o anterior ($)")
  plt.ylabel("Venta REMA ($)")
  plt.title("Regresi√≥n Lineal: Venta REMA vs Vta A√±o anterior",fontsize=14, fontweight='bold')
  plt.legend()
  plt.grid(alpha=0.3)
  plt.show()

# Llamar a la funci√≥n
plot(x, y, y_pred)
```

## Vemos que en muchos casos los valore predichos son muy cercanos a los reales. Se puede evidenciar en el grafico y al ver el dataset con la columna agregada de "**PredictedVentaREMA($)**"  <br> Ahora podemos calcular el error con las metricas correspondientes:

```python
print("R2=", r2_score(data["Venta REMA ($)"], y_pred))
mse = mean_squared_error(data["Venta REMA ($)"], y_pred)
rmse = np.sqrt(mse)
print("RMSE=", rmse)
```

### **1-Residuos sin estructira y media a cero**

```python
residuals = y - y_pred
print("Media de residuos: ", np.mean(residuals))
plt.scatter(x, residuals)
plt.axhline(y=0.0, color='r', linestyle='-')
plt.ylabel('Residuos',fontweight='bold')
plt.xlabel('variable independiente (Vta A√±o anterior)',fontweight='bold')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

## Se ve que la distribucion es bastante aleatoria

```python


# QQ-Plot para evaluar normalidad
sm.qqplot(residuals, line='45', fit=True)
plt.show()
```

### **3- Test Shapiro-Wilk (normalidad)**:<br>
### La hipotesis nula H0 es que la muestra sigue una distribucion normal (aceptada si p-value>0.05). La hipotesis alternativa Ha es que la muestra no viene de una distribucion normal. Si se rechaza H0, entonces los residuos no son normales.

```python
# Aplicamos el Shapiro-Wilk test
from scipy.stats import shapiro
shapiro(residuals)      #P-valor > 0.05 es (0.43) Los datos tienen una distribuci√≥n normal
```

### **3- Homocedasticidad**.
### Esto significa que la varianza se mantiene constante.<br>
**Test Breusch-Pagan (homoscedasticity)**:
### La hipotesis nula H0 es que existe homoscedasticity. La hipotesis alternativa Ha es que no hay homoscedasticity (sino que hay heteroscedasticity). Si p-value>=0.05, no podemos rechazar la hipotesis nula.

```python
# Aplicamos Bresuch-Pagan test
names = ['Lagrange multiplier statistic', 'p-value','f-value', 'f p-value']
test = smf.het_breuschpagan(model.resid, model.model.exog)
lzip(names, test)

#p-value', 0.06511058291376648  No se pude rechazar la H0 efectivamente hay Homoscedasticidad
```

### **4- Autocorrelacion entre los residuos**.
### Esto significa que que el valor de un residuo condiciona el valor de los residuos cercanos.<br>
### **Test de Durbin-Watson (autocorrelacion)**: La hipotesis nula H0 es que no existe correlacion entre los residuos. La hipotesis alternativa Ha es que los residuos est√°n autocorrelacionados.

El estadistico del test est√° en el rango [0-4], con la siguiente interpretacion:
* Si el estadistico est√° cercano a 0, hay mucha evidencia de correlacion serial positiva
* Si el estadistico est√° cercano a 4, hay mucha evidencia de correlacion serial negativa
* Normalmente, si el valor estadistico est√° entre 1.5 y 2.5 se considera normal. Si los valores est√°n fuera de ese rango, indicar√≠a que existen problemas de autocorrelacion


```python
# Aplicamos el Durbin-Watson test
durbin_watson(model.resid)  #--> Devuelve 2.011823328050704 Lo que indica que esta dentro de los valores normales
```
